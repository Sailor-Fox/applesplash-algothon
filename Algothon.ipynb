```python
x = range(250)
y = prices[:,range(50)]
plt.plot(x, y, label = range(50))
leg = plt.legend(bbox_to_anchor=(1,1), ncol = 5)
plt.title('prices for stocks 1-50')
plt.xlabel('days')
plt.ylabel('price')
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(prices)
print(df)
```

```python
diff = df.diff()
print("difference in daily price of stocks 0-49: ")
print(diff)

diff.to_csv(r'diff.csv')

average = abs(diff).mean()
print("average in abs difference in daily price of stocks 0-49: ")
print(average)

average.to_csv(r'average.csv')

import statistics as statistics

sd = statistics.stdev(abs(diff))
print("standard deviation")
print(sd)
```
